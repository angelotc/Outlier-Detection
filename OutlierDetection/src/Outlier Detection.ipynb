```python
import os
import numpy as np
import pandas as pd

def add_row(df, row):
    df.loc[-1] = row
    df.index = df.index + 1  
    return df.sort_index()

def main():
    
    #### Change directory to input folder. Make sure you copy the Zls to the 
    #### input folder before running the code.
    os.chdir("..")
    directory = os.path.abspath(os.getcwd())
    input_path = os.path.abspath((os.path.join(directory,'input')))
    output_path = os.path.abspath((os.path.join(directory,'output')))
    
    #### Create a dictionary
    #### {Cust # : lrrCount}
    customers = {}

    #### Loop through every file in the directory
    for filename in os.listdir(input_path):
        if filename.endswith(".txt"):
            try:
                #### Process files that end with .txt
                file = open(os.path.join(input_path, filename))
                lines = file.read()
                
                #print(lines)

                dl = '|'
                lines = lines.split(dl)
                lrr_count = lines[1]

                #print(filename)
                #print(lrr_count + '\n' )
                
                ### Populate our dictionary to be
                #### {cust # : lrr_count}
                customers[filename] = int(lrr_count)
                count_list.append(int(lrr_count))
                
                file.close()

            except:
                pass
        
    mean   =  np.mean(list(customers.values()))
    median =  np.median(list(customers.values()))
    #print('Mean:   ', mean)
    #print('Median: ', median)

    
    # Create output file with flagged customers
    df = pd.DataFrame(columns = ['Customer Name', 'Count', 'Median', 'Mean' ])
    for cust in customers:
        lrrCount = customers[cust]
        threshold = 10 # can change this as needed
        if (lrrCount >= (mean * threshold)) \
        or (lrrCount <= (mean / threshold)) \
        or (lrrCount >= (median * threshold)) \
        or (lrrCount <= (median / threshold)):
            row = [cust.split('.txt')[0], customers[cust], median, mean]
            add_row(df, row )
    
    # Create a CSV file
    df.to_csv(os.path.join(output_path, 'Affected Customers.csv'), index = False)
    
main()
```
